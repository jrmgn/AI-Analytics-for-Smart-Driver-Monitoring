```python
import pandas as pd
import numpy as np
import os

os.makedirs('../data', exist_ok=True)

np.random.seed(42)
N = 1000 

drivers = np.random.choice([f"D{d:03d}" for d in range(50)], N)
trip_ids = [f"T{i:05d}" for i in range(N)]

speeds = np.clip(np.random.normal(40, 15, N), 10, 140) 
hard_brake = (np.random.rand(N) < 0.07).astype(int)

ratings = (5.5 - (speeds/45) - (hard_brake * 1.5) + np.random.normal(0, 0.3, N))
ratings = np.clip(ratings, 1, 5).round().astype(int)

df_telem = pd.DataFrame({
    'trip_id': trip_ids,
    'driver_id': drivers,
    'avg_speed_kmh': speeds,
    'hard_brake_event': hard_brake,
    'rating': ratings
})

df_telem.to_csv("../data/telemetry_samples.csv", index=False)
df_telem.head()
```

```python
feedback_map = {
    5: ["Amazing driver, felt very safe.", "Smooth ride and professional.", "Clean car, great experience!"],
    4: ["Good trip, arrived on time.", "Steady driving, no complaints.", "Professional service."],
    3: ["Fine ride, but the car smelled a bit.", "Average experience.", "Driver was a bit slow."],
    2: ["Sudden braking made me nervous.", "Driver was speeding on the highway.", "A bit rude when I asked a question."],
    1: ["Terrible! Driver was reckless.", "Very unsafe driving, I was scared.", "Horrible experience, never again."]
}

df_telem['feedback_text'] = df_telem['rating'].apply(lambda x: np.random.choice(feedback_map[x]))
```
